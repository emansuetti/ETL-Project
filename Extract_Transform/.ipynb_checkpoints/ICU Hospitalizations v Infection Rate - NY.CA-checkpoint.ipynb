```python
NY_Data = pd.read_csv('../resources/ny_data.csv', index_col=0)
NY_Data.fillna(0, inplace=True)
NY_Data =NY_Data.drop(['test results total', 'test increase', 'positive increase'], axis = 1)
NY_Data
```

```python
CA_Data = pd.read_csv('../resources/clean_ca_data.csv', index_col=0)
CA_Data.fillna(0, inplace=True)
CA_Data = CA_Data.drop(['test results total', 'test increase', 'positive increase'], axis = 1)
CA_Data
```

```python
InfectionRate_ICU_Data = NY_Data.merge(CA_Data, on='date')

InfectionRate_ICU_Data
```

```python
InfectionRate_ICU_Data = InfectionRate_ICU_Data.rename(columns= {"date":"Date",
                                                                "deaths_x":"Deaths NY",
                                                                "Icu hospitalized_x":"ICU Hospitalizations NY",
                                                                "positive cases viral_x": "Positive Viral Cases NY",
                                                                "deaths_y":"Deaths CA",
                                                                "Icu hospitalized_y":"ICU Hospitalizations CA",
                                                                "positive cases viral_y": "Positive Viral Cases CA"
                                                                })
InfectionRate_ICU_Data = InfectionRate_ICU_Data.drop(['state_x', 'state_y'], axis = 1)
InfectionRate_ICU_Data
```

```python
InfectionRate_ICU_Data = InfectionRate_ICU_Data.reindex(columns=['Date', 'Deaths NY', 'Deaths CA', 'Positive Viral Cases NY', 'Positive Viral Cases CA', 'ICU Hospitalizations NY', 'ICU Hospitalizations CA'])
InfectionRate_ICU_Data
```
